```python
combination = data.groupby(['Member_number', 'Date'])['itemDescription'].apply(list)
from collections import Counter
from itertools import combinations
```

```python
d  = Counter()
a = combination.values
for sub in a:
    if len(a) < 2:
        continue
    sub.sort()
    for comb in combinations(sub,2):
        d[comb] += 1
```

```python
d.most_common()
```

```python
data['Date'] = pd.to_datetime(data['Date'])
shopping_times = data.drop('itemDescription', axis=1)
member_id = data['Member_number'].unique()
store = pd.DataFrame()
for i in member_id:
    temp = shopping_times[shopping_times['Member_number'] == i]['Date'].unique()
    month = []
    day_of_week = []
    for h in temp:
        conv = pd.to_datetime(h)
        month.append(conv.month)
        day_of_week.append(conv.weekday())
    result = pd.DataFrame({'month':month, 'day_of_week':day_of_week})
    store = pd.concat([store, result])
```

```python
store = store.reset_index(drop=True)
store
```

```python
for i in range(1, 13):
    print(store[store['month']==i].value_counts(sort=False))
```
